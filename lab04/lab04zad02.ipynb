Zaimportowanie potrzebnych bibliotek


```python
import matplotlib.pyplot as plt
import random
from aco import AntColony
```

Wyznaczenie przykładowych punktów do wyznaczenia między nimi optymalnej ścieżki


```python
COORDS = (
    (20, 52), (43, 50), (20, 84), (70, 65),
    (29, 90), (87, 83), (73, 23), (20, 45),
    (10, 0), (0, 47), (24, 54), (4, 23),
    (23, 87), (45, 76), (24, 87), (1, 89),
    (65, 92), (61, 69), (11, 77), (76, 12),
    (86, 12), (2, 1), (26, 12), (6, 12)
)
```

Stworzenie funkcji, która zwraca losowy punkt


```python
def random_coord():
    r = random.randint(0, len(COORDS))
    return r
```

Stworzenie funkcji rysujących punkty i wszystkie możliwe połączenia między nimi


```python
def plot_nodes(w=12, h=8):
    for x, y in COORDS:
        plt.plot(x, y, "g.", markersize=15)
    plt.axis("off")
    fig = plt.gcf()
    fig.set_size_inches([w, h])


def plot_all_edges():
    paths = ((a, b) for a in COORDS for b in COORDS)

    for a, b in paths:
        plt.plot((a[0], b[0]), (a[1], b[1]))
```

Narysowanie punktów i wszystkich możliwych połączeń


```python
plot_nodes()
plot_all_edges()
```

Uruchomienie algorytmu ACO dla podanych punktów z następującymi parametrami:

- ilość mrówek: `200`
- $\alpha$: `0.4`
- $\beta$: `1.5`
- szybkość odparowywania feromonu: `0.6`
- stała feromonu: `900.0`
- ilość iteracji: `300`

oraz wyświetlenie znalezionej optymalnej ścieżki między punktami


```python
colony = AntColony(COORDS, ant_count=200, alpha=0.4, beta=1.5,
                   pheromone_evaporation_rate=0.60, pheromone_constant=900.0,
                   iterations=300)

optimal_nodes = colony.get_path()

for i in range(len(optimal_nodes) - 1):
    plt.plot(
        (optimal_nodes[i][0], optimal_nodes[i + 1][0]),
        (optimal_nodes[i][1], optimal_nodes[i + 1][1]),
    )

plot_nodes()
plt.show()
```

Na szybkość znalezienia rozwiązania głównie wpływa ilość mrówek oraz ilość iteracji, w mniejszym stopniu także parametry $\alpha$ oraz $\beta$
