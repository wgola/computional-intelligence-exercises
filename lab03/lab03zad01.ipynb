Zaimportowanie bibliotek


```python
import numpy as np
import pygad
import math
```

Ustalenie przestrzeni genów jako liczb zmiennoprzecinkowych z zakresu `<0,1)` z dokładnością `0.0001`


```python
gene_space = np.arange(0, 1, 0.0001)
```

Stworzenie funkcji obliczającej wytrzymałość stopu oraz funkcji obliczającej na jej podstawie wyniku rozwiązania


```python
def endurance(x, y, z, u, v, w):
    return math.exp(-2 * (y-math.sin(x))**2)+math.sin(z*u)+math.cos(v*w)


def fitness_function(solution, solution_idx):
    return endurance(*solution)
```

Ustalenie wielkości populacji na `30` chromosomów oraz liczby genów każdego chromosmu na `6`


```python
sol_per_pop = 30
num_genes = 6
```

Ustalenie liczby rodziców do "rozmnażania" na `15`, liczby generacji na `50` oraz ilość rodziców do zachowania pomiędzy generacjami na `4`


```python
num_parents_mating = 15
num_generations = 50
keep_parents = 4
```

Ustalenie typu selekcji rodziców oraz ilości punktów, w których należy robić krzyżowanie


```python
parent_selection_type = "sss"
crossover_type = "single_point"
```

Ustalenie typu mutacji na losową oraz procentu mutowanych genów na `10%`


```python
mutation_type = "random"
mutation_percent_genes = 17
```

Inicjalizacja oraz uruchomienie algorytmu ze zdefiniowanymi parametrami


```python
ga_instance = pygad.GA(
    gene_space=gene_space,
    num_generations=num_generations,
    num_parents_mating=num_parents_mating,
    fitness_func=fitness_function,
    sol_per_pop=sol_per_pop,
    num_genes=num_genes,
    parent_selection_type=parent_selection_type,
    keep_parents=keep_parents,
    crossover_type=crossover_type,
    mutation_type=mutation_type,
    mutation_percent_genes=mutation_percent_genes
)

ga_instance.run()
```

Najlepszy znaleziony chromosom oraz jego ocena


```python
solution, solution_fitness, solution_idx = ga_instance.best_solution()
print("Parameters of the best solution : {solution}".format(solution=solution))
print("Fitness value of the best solution = {solution_fitness}".format(
    solution_fitness=solution_fitness))
```

Wykres prezentujący zmianę oceny na przestrzeni pokoleń


```python
ga_instance.plot_fitness()
print()
```
